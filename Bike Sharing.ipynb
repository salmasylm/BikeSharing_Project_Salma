## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
hourly_counts = hour.groupby(by=["hr", "yr"]).agg({
    "cnt": "sum"
}).reset_index()

plt.figure(figsize=(10, 6))

sns.lineplot(data=hourly_counts, x="hr", y="cnt", hue="yr", palette=["#FF6347", "#4682B4"], marker="o", linestyle="--")
sns.scatterplot(data=hourly_counts, x="hr", y="cnt", hue="yr", palette="coolwarm", marker="o")

plt.ylabel("Jumlah Penyewaan Sepeda")
plt.xlabel("Jam-jam harian dalam 2 tahun")
plt.title("Jumlah Total penyewaan sepeda berdasarkan Jam dan Tahun")
plt.legend(title="Year", loc="upper right")
plt.xticks(ticks=hourly_counts["hr"], labels=hourly_counts["hr"])
plt.tight_layout()
plt.show()
```

**Insight:**
Penyewaan sepeda menunjukkan pola harian yang sangat dipengaruhi oleh jam sibuk kerja. Terdapat dua puncak aktivitas yang signifikan, yaitu pada pukul 08.00 pagi dan pukul 17.00 hingga 18.00 sore. Hal ini mengindikasikan bahwa sepeda banyak digunakan sebagai alat transportasi komuter. Tahun 2012 mencatat volume penyewaan yang jauh lebih tinggi dibandingkan tahun 2011 di setiap jamnya. Aktivitas penyewaan berada di titik terendah pada dini hari antara pukul 00.00 sampai 05.00.

```python
bulanan_counts = day.groupby(by=["mnth", "yr"]).agg({
    "cnt": "sum"
}).reset_index()

plt.figure(figsize=(10, 6))

sns.lineplot(data=bulanan_counts, x="mnth", y="cnt", hue="yr", palette=["#FF6347", "#4682B4"], linestyle='--', marker="o")

plt.ylabel("Jumlah Penyewaan Sepeda")
plt.xlabel("Bulan")
plt.title("Jumlah penyewaan sepeda selama periode 2 tahun berdasarkan bulan")
plt.legend(title="Tahun", loc="upper right")
plt.xticks(ticks=bulanan_counts["mnth"], labels=bulanan_counts["mnth"])
plt.tight_layout()
plt.show()
```

**Insight:**
Tren bulanan memperlihatkan adanya pengaruh musim yang kuat terhadap jumlah penyewaan. Permintaan mulai meningkat pada bulan Maret dan mencapai puncaknya selama periode pertengahan tahun, yaitu antara bulan Juni hingga September. Setelah itu, jumlah penyewaan menurun secara konsisten hingga akhir tahun. Perbandingan tahun ke tahun menunjukkan pertumbuhan bisnis yang pesat, di mana performa pada tahun 2012 selalu berada di atas capaian tahun 2011 untuk seluruh bulan.

### Pertanyaan 2:

```python
jumlah_per_hari_kerja = day.groupby(by=["workingday", "yr"]).agg({
    "cnt": "sum"
}).reset_index()

sns.barplot(data=jumlah_per_hari_kerja, x="workingday", y="cnt", hue="yr", palette=["#FF6347", "#4682B4"])

plt.xlabel("Hari")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.title("Total Penyewaan Sepeda Berdasarkan Weekdays/Weekends")
plt.legend(title="Tahun", loc="upper right")

plt.xticks(ticks=jumlah_per_hari_kerja["workingday"].unique(), labels=["Weekends", "Weekdays"])

plt.tight_layout()
plt.show()
```

**Insight:**
Data menunjukkan bahwa jumlah penyewaan sepeda pada hari kerja (weekdays) secara konsisten jauh lebih tinggi dibandingkan pada akhir pekan (weekends) di kedua tahun tersebut. Pada hari kerja di tahun kedua, total penyewaan melonjak signifikan hingga melewati angka 1,4 juta, yang hampir dua kali lipat dari capaian tahun sebelumnya. Dominasi volume pada hari kerja ini memperkuat indikasi bahwa penggunaan sepeda didorong oleh kebutuhan transportasi rutin para pekerja untuk berkomuter. Meskipun volume di akhir pekan lebih rendah, tetap terlihat tren pertumbuhan yang positif dari tahun ke tahun yang menunjukkan peningkatan popularitas layanan secara keseluruhan.

## Analisis Lanjutan

Data harian dan bulanan menunjukkan sinergi kuat di mana perilaku komuter pada jam sibuk, yaitu pukul 08.00 dan 17.00 hingga 18.00, menjadi penggerak utama volume total penyewaan setiap bulannya. Pertumbuhan signifikan di tahun kedua memperlihatkan bahwa peningkatan ini terjadi secara merata di sepanjang hari dan sepanjang tahun, yang menandakan perluasan basis pengguna secara masif. Selain itu, pergeseran puncak bulanan dari bulan Juni pada tahun pertama ke bulan September pada tahun kedua menunjukkan bahwa layanan ini semakin optimal digunakan saat kondisi cuaca mendukung mobilitas luar ruangan.

Kesenjangan volume antara hari kerja dan akhir pekan membuktikan bahwa sepeda digunakan sebagai alat transportasi esensial untuk mendukung produktivitas kerja dibandingkan sekadar aktivitas rekreasi. Pertumbuhan di hari kerja (weekdays) yang mencapai angka total lebih dari 1.400.000 penyewaan pada tahun kedua menunjukkan efektivitas layanan dalam mengintegrasikan diri ke dalam infrastruktur transportasi harian masyarakat. Meskipun penggunaan di akhir pekan memiliki volume yang lebih rendah, peningkatan jumlah penyewaan yang stabil dari tahun ke tahun mencerminkan pertumbuhan kepercayaan pelanggan di seluruh segmen waktu.
